```python
f = open('26.txt', 'r')

n = int(f.readline())

arr = [int(line) for line in f.readlines()]

disc_n = len(arr) // 6

arr.sort()

print(f'один чек = {sum(arr[:disc_n]) // 2 + sum(arr[disc_n:])}')

sm = 0

bucket = []


for elem in sorted(arr, reverse=True):
    if len(bucket) < 6:
        bucket.append(elem)
    else:
        sm = sm + sum(bucket) - (min(bucket) // 2)
        bucket = [elem]

print(f'разные чеки = {sm + sum(bucket)}')
```

```python
f = open('26_4.txt', 'r')

n = int(f.readline())

sizes = list(set(map(int, f.readlines())))
sizes.sort()

#f[n] = размер самой длинной цепочки коробок, которая начинается с sizes[n]
#f[n] = 1 + f[n + 1], если sizes[n + 1] - sizes[n] >= 3
#f[n] = 1 + f[n + 2], если sizes[n + 2] - sizes[n] >= 3
#
#...
#
#f[n] = 1 + max(f[n + k], f[n + 2], ..., f[len(sizes) - 1]), где k -- такое минимальное число, что sizes[n + k] - sizes[n] >= 3

m = len(sizes)

f = [1] * m

f[len(sizes) - 1] = 1

for i in range(m - 2, -1, -1):
    k = 1
    while k + i < m:
        if (sizes[k + i] - sizes[i] >= 9) and (f[i] <= 1 + f[i + k]):
            f[i] = f[i + k] + 1

        k += 1
    
print(max(f))

for i in range(m-1, -1, -1):
    if f[i] == max(f): 
        print(sizes[i])
        break
```

```python
f = open('26_5.txt', 'r')

n = int(f.readline())

times = []
for line in f.readlines():
    times.append(tuple(map(int, line.split())))


#arr[i] = колво посетитетлей, которые были с i по i + 1 минуту в магазине

arr = [0] * 1440

for time in times:
    start = time[0]
    end = time[1]

    for j in range(start, end):
        arr[j] += 1


hl = '*'.join(list(map(str, arr)))

hl
```
