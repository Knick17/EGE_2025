```python
def ok(a):
    return a % 7 == 0 and abs(a) <= 999 and abs(a) >= 100

f = open('17-1.txt', 'r')

arr = [int(x) for x in f.readlines()]

mx = max([x for x in arr if x % 10 == 7])

cnt = 0
mx_sm = -1

for i in range(len(arr) - 2):
    a = arr[i]
    b = arr[i + 1]
    c = arr[i + 2]

    if str(abs(a))[0] == str(abs(b))[0] and str(abs(b))[0] == str(abs(c))[0]:
        if ok(a) or ok(b) or ok(c):
            if abs(a + b + c) < mx:
                cnt += 1
                mx_sm = max(mx_sm, abs(a + b + c))
```

```python
field = [['_'] * 77 for i in  range(77)]

#field[i][j] -- ситуация когда в перой куче j камней, во второй i камней

for i in range(77):
    for j in range(77):
        if j < 7 and j >= 1: field[i][j] = 'e'
        elif i == 0 or j == 0: field[i][j] = str(i + j)
        elif i + j >= 77: field[i][j] = 'x'

for i in range(77):
    for j in range(77):
        if field[i][j] == '_':
            if i * 2 + j >= 77 or i + j + 1 >= 77 or i + 2 * j >= 77: field[i][j] = 'w1'

for i in range(77):
    for j in range(77):
        if field[i][j] == '_':
            if field[i * 2][j] == 'w1' \
            and field[i][j * 2] == 'w1' \
            and field[i][j +1] == 'w1' \
            and field[i + 1][j] == 'w1': field[i][j] = 'l1'


w = open('wrt.csv', 'w')

w.writelines([','.join(elem) + '\n' for elem in field])
```

```python
def f(n):
    arr = [(n + n // k) for k in range(2, n + 1) if n % k == 0]
    return arr

# f(n) возвращает все возможные преобразования над кучей

def game(n):
    if any(x > 45 for x in f(n)): return 'w1'
    if all (game(x) == 'w1' for x in f(n)): return 'l1'
    if any (game(x) == 'l1' for x in f(n)): return 'w2'
    if all (game(x) =='w1' or game(x) == 'w2' for x in f(n)): return 'l2'
    return 'хз'


for n in range(1, 46):
    print(n, '\t', game(n))
```
