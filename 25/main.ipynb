```python
# ДЛИНА искомого числа не больше 10 цифр, 5 уже занято => нужно перебрать все возможные последовательнсти 
import itertools as it

arr = []
for k in range(6):  
    for i in it.product('0123456789', repeat=k):
        for j in it.product('0123456789', repeat=5-k):
            n = int('48' + ''.join(i) + '15' + ''.join(j) + '0')
            if n % 42 == 0 and n <= 10_000_000_000:
                arr.append(n)

for elem in sorted(arr, reverse=True)[:6]:
    print(elem // 42)


```
