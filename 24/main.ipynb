```python
f = open('24.txt', 'r')

min_G = 10e10
lines = f.readlines()

for line in lines:
    if line.count('G') <= min_G: min_G = line.count('G')

d = dict()

for line in lines:
    if line.count('G') == min_G:
        print(line)
        
        for word in line: d[word] = line.count(word)
        break

mx = max(d.values())
max([elem for elem in d.keys() if d[elem] == mx])


```

```python
f = open('24_2.txt', 'r')

st = f.readline().replace('X', '*').replace('Z', '*')

mx_len = -1
for l in st.split('*'): mx_len = max(mx_len, len(l))

print(mx_len)
```

```python
f = open('24_3.txt', 'r')
st = f.readline()
letters = 'QWERTYUIOPASDFGHJKLZXCVBNM'

for letter in letters:
    st = st.replace(letter, '*')

st = st.replace('\n', '').replace(' ', '')

for i in range(2, 20):
    st = st.replace('*' * i, '*')

print(st.split('*'))

mx = -1

for elem in st.split('*'):
    if elem != '': mx = max(int(elem), mx)

print(mx)
```

```python
f = open('24_5.txt', 'r')

s = f.readline().replace('8', '1').replace('D', 'R')

# idx[i] = true ЕСЛИ STR[i:i+2] = 11r иначе false

idx = [0] * len(s)

#11r

for i in range(len(s)- 2):
    if s[i:i+3] == '11R': idx[i] = 1

print(idx[:100])
mx = -1
cnt = 0
i = 0
while i < len(idx):
    if idx[i] == 1: 
        cnt += 1
        i += 3
    else:
        mx = max(mx, cnt)
        cnt = 0
        i += 1
    
print(mx)
```
